```julia
using PorousMaterials
```

```julia
crystals = ("SBMOF-1.cssr", "NiPyC2.cif",  "COF-103.cif")
```

```julia
for crystal in crystals
    println("Runnig mol sim in ", crystal)
    frame = Framework(crystal)
    strip_numbers_from_atom_labels!(frame)
    T = 298.0
    forcefield = LJForceField("UFF.csv", cutoffradius=14.0, mixing_rules="Lorentz-Berthelot")
    xe = Molecule("Xe")
    P = 10 .^ range(-2, stop=log10(300), length=15)
    n_sample_cycles = 25000
    n_burn_cycles = 25000
    adsorption_data = adsorption_isotherm(frame, xe, T, P, forcefield, 
                        n_burn_cycles=n_burn_cycles, n_sample_cycles=n_sample_cycles,
                        eos=:PengRobinson)
end
```
